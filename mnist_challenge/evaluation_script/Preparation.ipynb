```python
import numpy as np
import pandas as pd


train = pd.read_csv("Data/train.csv")
print(train)
```

```python
X = train.loc[:, train.columns != 'label']
Y = train.loc[:, train.columns == 'label']
X = X.as_matrix()
Y = Y.as_matrix()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
z = pd.DataFrame(X_train)
z['label'] = y_train
z = z.set_index('label').reset_index()
```

```python
z.to_csv("Data/training.csv")
```

```python
zt = pd.DataFrame(X_test)
zt2 = pd.DataFrame(y_test)
zt3 = zt2.sample(frac=1)
zt3.reset_index(inplace=True)
zt3.drop("index",axis=1,inplace=True)
zt3 = zt3.rename(index=str, columns={0: "label"})
zt2 = zt2.rename(index=str, columns={0: "label"})
```

```python
zt3.to_csv("Data/submission.csv")
```

```python
zt2.to_csv("Data/answers.csv")
```

```python
zt.to_csv("Data/testing.csv")
```
